```python
from bs4 import BeautifulSoup as bs
import requests
from selenium import webdriver
from splinter import Browser
import time
```

```python
browser = Browser('chrome', headless=False)
```

```python
url = 'https://twitter.com/marswxreport?lang=en'
```

```python
browser.visit(url)
```

```python
time.sleep(10)
html = browser.html
soup = bs(html, 'html.parser')
```

```python
print(soup.prettify())
```

```python
temp = soup.find_all(class_="css-1dbjc4n")[0]
```

```python
list_info = temp.find_all('span', class_="css-901oao css-16my406 r-1qd0xha r-ad9z0x r-bcqeeo r-qvutc0")
```

```python
for element in list_info:
    if element.text.find("InSight")!=-1:
        mars_weather = element.text
        break
```
